```python
train_join_df = pd.read_csv("../logs/mitbaseline_train_acc_class_w_h.csv")
train_join_df.head(2)
```

```python
Scene_acc = train_join_df.groupby(["Scence"]).mean()["Accuracy"]
```

```python
plt.plot(Scene_acc.sort_values().values)
```

```python
Scene_acc.sort_values()
```

```python
y = train_join_df["Accuracy"].values
X = train_join_df.drop(["Pic", "Scence", "Accuracy"], axis=1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
columns = train_join_df.drop(["Pic", "Scence", "Accuracy"], axis=1).columns
```

```python
rfr = RFR(n_estimators=100, n_jobs=-1)
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```

```python
def plot_feat_importance(feature_importances,index):
    feat_scores = pd.DataFrame({"Mean Decrease Impurity" : feature_importances},
    index=index)
    feat_scores = feat_scores.sort_values(by="Mean Decrease Impurity")
    feat_scores.plot(kind="barh", figsize=(10,30))
plot_feat_importance(rfr.feature_importances_,columns)
```

```python
def percentile_avg(df, col, col_avg, num_perc = 10):
    percentile = df.loc[df[col] >0, col].quantile(np.linspace(0, 1.1, 11, 0))
    percentile = percentile.unique()
    x_tick = range(len(percentile))
    x_label = []
    avg_col = []
    err_col = []
    min_col = []
    max_col = []
    for i in range(1, len(x_tick)):
#         print(col, (df[(df[col] >= percentile[i-1]) & (df[col] < percentile[i])].count()["Accuracy"]))
        avg_col.append(df[(df[col] >=percentile[i-1]) & (df[col] < percentile[i])].mean()["Accuracy"])
        err_col.append(df[(df[col] >=percentile[i-1]) & (df[col] < percentile[i])].var()["Accuracy"])
        min_col.append(df[(df[col] >=percentile[i-1]) & (df[col] < percentile[i])].min()["Accuracy"])
        max_col.append(df[(df[col] >=percentile[i-1]) & (df[col] < percentile[i])].max()["Accuracy"])
        x_label.append(str((round((percentile[i-1] + percentile[i])/2,2))))
    x = range(len(x_tick)-1)
    print(max_col)
    plt.bar(x, avg_col, yerr=[[t1 - t2 for t1, t2 in zip(avg_col, min_col)],
                              [t1 - t2 for t1, t2 in zip(max_col, avg_col)]])
    plt.xticks(x, x_label)
    plt.title(col)
percentile_avg(train_join_df, "class0", "Accuracy")
plt.show()
percentile_avg(train_join_df, "class2", "Accuracy")
plt.show()
percentile_avg(train_join_df, "class3", "Accuracy")
plt.show()
percentile_avg(train_join_df, "class4", "Accuracy")
plt.show()
percentile_avg(train_join_df, "class24", "Accuracy")
plt.show()
# percentile_avg(train_join_df, "H", "Accuracy")
# plt.show()
# percentile_avg(train_join_df, "W", "Accuracy")
# plt.show()
```
